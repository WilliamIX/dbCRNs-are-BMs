```python
import numpy as np
import pylab as plt
import os


epsilon = 0.0001
E = np.array([[0, 1.0], [1.0, 0]])+epsilon
E = E/np.sum(E)

def DKL(Q, P):
    tot = 0
    tot += Q[0, 1]*np.log(Q[0, 1]/P[0, 1])+Q[1, 0]*np.log(Q[1, 0]/P[1, 0])
    tot += Q[0, 0]*np.log(Q[0, 0]/P[0, 0])+Q[1, 1]*np.log(Q[1, 1]/P[1, 1])
    return tot


uniform = np.ones((2, 2))
uniform = uniform/np.sum(uniform)

delta_inv = [2, 8, 32, 128, 512, 2048]

path = os.getcwd()+"/Simulation Arrays/Run "

DKL_QX = {}
DKL_EQ = {}
DKL_EX = {}
dSdt = {}

DKL_QXf = np.zeros((6, 6))
DKL_EQf = np.zeros((6, 6))
DKL_EXf = np.zeros((6, 6))


DQf = {}
DXf = {}
DEf = {}

iterations = [i for i in range(0, 40, 1)]
plot_iteration = [i for i in range(0, 41, 1)]

ind_l = 8
plt.figure(figsize = (12, 16))

runs = [0, 1, 2]
for ind, di_l in enumerate([2, 8, 32, 128, 512, 2048]):#, 8, 32, 128, 512]:
    for ind2, di_c in enumerate([2, 8, 32, 128, 512, 2048]):
        
        DKL_QX[di_l, di_c] = np.zeros(len(iterations)+1)
        DKL_EX[di_l, di_c] = np.zeros(len(iterations)+1)
        DKL_EQ[di_l, di_c] = np.zeros(len(iterations)+1)
        
        DKL_QX[di_l, di_c][0] = DKL(uniform, uniform)
        DKL_EX[di_l, di_c][0] = DKL(E, uniform)
        DKL_EQ[di_l, di_c][0] = DKL(E, uniform)
        
        
        DQf[di_l, di_c] = np.zeros((2, 2))
        DXf[di_l, di_c] = np.zeros((2, 2))
        
        #dSdt[di_l, di_c] = np.zeros(len(iterations))
        
        for run in runs:
            
            for iteration in iterations:
                print("loading ","run", run, f"(l, c, i) = ({di_c}, {di_l}, {iteration})", end = "...")
                
                #dsdt = np.load(path+f"{run}/distQ_dcopy={di_c}_dlearn={di_l}_iter={iteration}.npy")/(1.0*len(runs))
                #raise RuntimeError
                #dSdt[di_l, di_c][iteration] += np.load(path+f"{run}/distQ_dcopy={di_c}_dlearn={di_l}_iter={iteration}.npy")/(1.0*len(runs))
                
                DQ = np.load(path+f"{run}/distQ_dcopy={di_c}_dlearn={di_l}_iter={iteration}.npy") + epsilon
                DQ = DQ/np.sum(DQ)
                #print("DKL_E(DQ)", DKL_E(DQ), end = "...")
                DKL_EQ[di_l, di_c][iteration+1] += DKL(E, DQ)/(1.0*len(runs))

                DX = np.load(path+f"{run}/distX_dcopy={di_c}_dlearn={di_l}_iter={iteration}.npy") + epsilon
                DX = DX/np.sum(DX)
                #print("DKL_E(DX)", DKL_E(DX), end = "...")
                DKL_EX[di_l, di_c][iteration+1] += np.array(DKL(E, DX))/(1.0*len(runs))

                DKL_QX[di_l, di_c][iteration+1] += np.array(DKL(DQ, DX))/(1.0*len(runs))

                if iteration in iterations[-1:]:
                    DQf[di_l, di_c] += DQ/(1.*len(runs))
                    DXf[di_l, di_c] += DX/(1.*len(runs))


            DKL_EQf[ind, ind2] += DKL(E, DQf[di_l, di_c])/(1.*len(runs))
            DKL_EXf[ind, ind2] += DKL(E, DXf[di_l, di_c])/(1.*len(runs))
            DKL_QXf[ind, ind2] += DKL(DQf[di_l, di_c], DXf[di_l, di_c])/(1.*len(runs))

            plt.subplot(6, 3, 3*ind+1)
            plt.pcolor(E)
            plt.subplot(6, 3, 3*ind+2)
            plt.pcolor(DQf[di_l, di_c])
            plt.title(f"di_c={di_c} di_l={di_l}")
            plt.subplot(6, 3, 3*ind+3)
            plt.pcolor(DXf[di_l, di_c])
        
        
print("Done")
```
