```python
%matplotlib inline
#Bioscrape can be installed with pip install bioscrape. For more details on installation, visit the Github page.
from bioscrape.types import Model
from bioscrape.simulator import py_simulate_model
import numpy as np
import pylab as plt
from scipy.sparse import csc_matrix
import matplotlib
from time import process_time
import itertools

from dbCRNfuncs import *
```

# Learn a Product Poisson 1 vesicle, stochastic implementation

```python
#Stochastic learning of a Product Poisson
k_learn = 1.0
kdb = 10
dt = 1.
EX = -3
EP = 0
Q0 = 100
P0 = 1000
X0 = 2
tmax = 10000
timepoints = np.arange(0, tmax, dt)

#Learning CRN
opposite_side_potential = False
both_side_potentials = False
assert both_side_potentials*opposite_side_potential is not True

species = ["X", "Q", "P"]
E = {"X":EX, "Q":0, "P":EP}
x0 = {"X":X0, "Q":Q0, "P":P0}
reactions = []
if opposite_side_potential:
    reactions += db_rxns(["P"], ["X"], E, k = kdb)
elif both_side_potentials:
    E["Po"] = EP
    x0["Po"] = P0
    reactions += db_rxns(["Po"], ["X", "P"], E, k = kdb)

else:
    reactions += db_rxns([], ["X", "P"], E, k = kdb)

reactions += clamping_reactions("P", "Q", "X", epsilon = k_learn, opposite_side_potential = opposite_side_potential)
if both_side_potentials:
    reactions += clamping_reactions("Po", "Q", "X", epsilon = k_learn, opposite_side_potential = not opposite_side_potential)
    
M_pp = Model(species = species, reactions = reactions, initial_condition_dict = x0)


results = py_simulate_model(timepoints, M_pp, stochastic = True)

Pf = np.mean(results["P"].to_numpy()[int(len(timepoints)/2):])
Exf = EX + np.log(Pf)


M_pp_free = Model(species = ["X", "Q"], reactions = db_rxns([], ["X"], E, k = kdb), initial_condition_dict = x0)
R_free = py_simulate_model(timepoints, M_pp_free, stochastic = True)

E_clamped = {"X":Exf, "P":0, "Q":0}
M_pp_clamped = Model(species = ["X", "Q"], reactions = db_rxns([], ["X"], E_clamped, k = kdb), initial_condition_dict = x0)
results_clamped = py_simulate_model(timepoints, M_pp_clamped, stochastic = True)

plt.figure(figsize = (8, 5))
plt.subplot(211)

max_ind = int(len(timepoints)/10)
title = "\nFree CRN: $\emptyset \leftrightharpoons S$\n"
title += "Clamped CRN: $\emptyset \leftrightharpoons S + P \quad P + S \\rightarrow 2P + S \quad Q + P \\rightarrow Q$"
plt.title(title, fontsize = "xx-large", loc = 'left')
plt.plot(R_free["time"][:max_ind], R_free["X"][:max_ind], color = 'cyan', label = "S Free")
plt.plot(results["time"][:max_ind], results["Q"][:max_ind], color = 'lightgreen', label = "Q", lw = 3)
plt.plot(results["time"][:max_ind], results["X"][:max_ind], color = 'blue', label = "S Clamped")
plt.plot(results["time"][:max_ind], results["P"][:max_ind], color = 'orange', label = "P")
plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("counts trajectory")
plt.yscale('log')




plt.subplot(212)
bins = int(max(results["X"].to_numpy())+1)
plt.hist(R_free["X"].to_numpy()[int(len(timepoints)/2):], bins = bins, range = (0, bins), density = True, color = 'cyan');
plt.hist(results["X"].to_numpy()[int(len(timepoints)/2):], bins = bins, range = (0, bins), density = True, color = 'blue');

plt.axvline(Q0, lw = 3, color = "lightgreen")

meanX_free = np.mean(R_free["X"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanX_free, label = "$<S>_{free}$", ls = "--", color = "gray")

meanX_clamped = np.mean(results["X"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanX_clamped, label = "$<S>_{clamped}$", color = "gray")

plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("count probability")
plt.tight_layout()
#plt.savefig("clamping_fig.pdf")
```

# Learn a Poisson N vesicles stochastic implementation

```python
#Stochastic Learning wth N Vesicles
N = 10 #Number of vesicles
V = N #Large Volume
v = N/V #Vesicle Concentration

k = k_learn#(k_learn*1.0/N, 1.0*k_learn*v) #rescale learning rate to match vesicle concentration

#Learning CRN
opposite_side_potential = False


species = ["P", "Q"]
reactions = []
E = {"P":EP, "Q":0}
x0 = {"P":P0, "Q":Q0}
for i in range(N):
    X = "X"+str(i)
    species.append(X)
    E[X]=EX
    x0[X]=X0
    
    if opposite_side_potential:
        reactions += db_rxns(["P"], [X], E, k = kdb)
    else:
        reactions += db_rxns([], [X, "P"], E, k = kdb) 
        
    reactions += clamping_reactions("P", "Q", X, epsilon = k_learn, k = k, V = V,
                                    opposite_side_potential = opposite_side_potential)

M_pp_N = Model(species = species, reactions = reactions, initial_condition_dict = x0)
results = py_simulate_model(timepoints, M_pp_N, stochastic = True)

plt.figure(figsize = (8, 5))
plt.subplot(211)

title = "\nFree CRN: $\emptyset \leftrightharpoons S$\n"
title += "Clamped CRN: $\emptyset \leftrightharpoons S + P \quad P + S \\rightarrow 2P + S \quad Q + P \\rightarrow Q$"
plt.title(title, fontsize = "xx-large", loc = 'left')
#plt.plot(R_free["time"][:max_ind], R_free["X"][:max_ind], color = 'cyan', label = "S Free")
plt.plot(results["time"][:max_ind], results["Q"][:max_ind], color = 'lightgreen', label = "Q", lw = 3)
for i in range(N):
    if i == N/2:
        label = "S Clamped"
    else:
        label = None
    plt.plot(results["time"][:max_ind], results[f"X{i}"][:max_ind], color = (0, 0, .5+.5*i/N), label = label, alpha = .5)
plt.plot(results["time"][:max_ind], results["P"][:max_ind], color = 'orange', label = "P")
plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("counts trajectory")
plt.yscale('log')




plt.subplot(212)
all_results = np.array([])
for i in range(N):
    all_results = np.concatenate((all_results, results[f"X{i}"][int(len(timepoints)/2):].to_numpy()))

meanX_clamped = np.mean(all_results)
bins = int(max(all_results)+1)
plt.hist(R_free["X"].to_numpy()[int(len(timepoints)/2):], bins = bins, range = (0, bins), density = True, color = 'cyan');
plt.hist(all_results, bins = bins, range = (0, bins), density = True, color = 'blue');

plt.axvline(Q0, lw = 3, color = "lightgreen")

#meanX_free = np.mean(R_free["X"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanX_free, label = "$<S>_{free}$", ls = "--", color = "gray")

#meanX_clamped = np.mean(results["X"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanX_clamped, label = "$<S>_{clamped}$", color = "gray")

plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("count probability")
plt.tight_layout()
```

# Learn a Product Poisson N Vesicles Hybrid Model Q Varying

```python
#Hybrid learning a Product Poisson with Q = varying.

#Learning CRN
species = ["P", "Q"]
E = {"X":EX, "P":EP, "Q":0}
x0 = {"X":X0, "P":P0, "Q":Q0}
reactions = []
for i in range(N):
    X = "X"+str(i)
    E[X] = EX
    x0[X] = X0
    species += [X]
    reactions += chemostat_db_rxns([], [X], [], ["P"], E, k = kdb)
    
rules = []
for i in range(N):
    X = "X"+str(i)
    rules.append(create_rule_tuple("P", "Q", X, k_learn, dt))

M = Model(species = species, reactions = reactions, initial_condition_dict = x0, rules = rules)


t_list = [0, 10]
Q_list = [Q0, Q0/10]
equation = f"Q = "
for i in range(len(t_list)):
    t = t_list[i]
    Q = Q_list[i]
    if i < len(t_list)-1:
        tt = t_list[i+1]
        equation += f"{Q}*Heaviside(t-{t})*(1-Heaviside(t-{tt}))+"
    else:
        equation += f"{Q}*Heaviside(t-{t})"
    
M.create_rule("assignment", {"equation":equation})

results_track = py_simulate_model(timepoints, M, stochastic = True)

Pf = np.mean(results_track["P"].to_numpy()[int(len(timepoints)/2):])
Exf = EX + np.log(Pf)


M_free_det = Model(species = ["X", "Q"], reactions = db_rxns([], ["X"], E, k = kdb), initial_condition_dict = x0)
R_free = py_simulate_model(timepoints, M_free_det, stochastic = True)

E_clamped = {"X":Exf, "P":0, "Q":0}
M_clamped_det = Model(species = ["X", "Q"], reactions = db_rxns([], ["X"], E_clamped, k = kdb), initial_condition_dict = x0)
R_clamped = py_simulate_model(timepoints, M_clamped_det, stochastic = True)

plt.figure(figsize = (8, 5))
plt.subplot(211)

title = "\nFree CRN: $\emptyset \leftrightharpoons S$\n"
title += "Clamped CRN: $\emptyset \leftrightharpoons S + P \quad P + S \\rightarrow 2P + S \quad Q + P \\rightarrow Q$"
plt.title(title, fontsize = "xx-large", loc = 'left')
plt.plot(R_free["time"], R_free["X"], color = 'cyan', label = "S Free")
for i in range(N):
    if i == int(N/2): label = "S Clamped"
    else: label = None
        
    plt.plot(results_track["time"], results_track[f"X{i}"], color = (0, 0, .5+.5*i/N), label = label, alpha = .5)

plt.plot(results_track["time"], results_track["Q"], color = 'lightgreen', label = "Q", lw = 3)
plt.plot(results_track["time"], results_track["P"], color = 'orange', label = "P")
plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("counts trajectory")
plt.yscale('log')

plt.subplot(212)
all_results_track = np.array([])
for i in range(N):
    all_results_track = np.concatenate((all_results_track, results_track[f"X{i}"].to_numpy()[int(len(timepoints)/10):]))
    
bins = int(max(all_results)+1)
plt.hist(R_free["X"].to_numpy()[int(len(timepoints)/2):], bins = bins, range = (0, bins), density = True, color = 'cyan');

plt.hist(all_results_track, bins = bins, range = (0, bins), density = True, color = 'blue');

plt.axvline(Q0, lw = 3, color = "lightgreen")
plt.axvline(Q0/10, lw = 3, color = "lightgreen")

meanX_free = np.mean(R_free["X"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanX_free, label = "$<S>_{free}$", ls = "--", color = "gray")

meanX_clamped = np.mean(all_results_track)
plt.axvline(meanX_clamped, label = "$<S>_{clamped}$", color = "gray")

plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("count probability")
plt.tight_layout()
```

# Learn a Product Poisson N Vesicles Hybrid Model Q Stochastic

```python
#Hybrid learning a Product Poisson with Q stochastic.
N = 10
kdb = 500
dt = .001
k_learn = .001
k_learn = (k_learn, k_learn)


#Learning CRN
species = ["P", "Q"]
E = {"X":EX, "P":EP, "Q":-4}
x0 = {"X":X0, "P":P0, "Q":Q0}
reactions = []
for i in range(N):
    X = "X"+str(i)
    E[X] = EX
    x0[X] = X0
    species += [X]
    reactions += chemostat_db_rxns([], [X], [], ["P"], E, k = kdb)

reactions += db_rxns([], ["Q", "Q"], E, k = kdb)

rules = []
for i in range(N):
    X = "X"+str(i)
    rules.append(create_rule_tuple("P", "Q", X, k_learn, dt))

M = Model(species = species, reactions = reactions, initial_condition_dict = x0, rules = rules)
    
timepoints = np.arange(0, 12, dt)
results_dist = py_simulate_model(timepoints, M, stochastic = True)

Pf = np.mean(results["P"].to_numpy()[int(len(timepoints)/2):])
Exf = EX + np.log(Pf)


M_free_det = Model(species = ["X", "Q"], reactions = db_rxns([], ["X"], E, k = kdb), initial_condition_dict = x0)
R_free = py_simulate_model(timepoints, M_free_det, stochastic = True)

E_clamped = {"X":Exf, "P":0, "Q":0}
M_clamped_det = Model(species = ["X", "Q"], reactions = db_rxns([], ["X"], E_clamped, k = kdb), initial_condition_dict = x0)
R_clamped = py_simulate_model(timepoints, M_clamped_det, stochastic = True)

plt.figure(figsize = (8, 5))
plt.subplot(211)

#max_ind = int(len(timepoints)/50)
title = "\nFree CRN: $\emptyset \leftrightharpoons S$\n"
title += "Clamped CRN: $\emptyset \leftrightharpoons S + P \quad P + S \\rightarrow 2P + S \quad Q + P \\rightarrow Q$"
plt.title(title, fontsize = "xx-large", loc = 'left')
plt.plot(R_free["time"], R_free["X"], color = 'cyan', label = "S Free")
for i in range(N):
    if i == int(N/2): label = "S Clamped"
    else: label = None
        
    plt.plot(results_dist["time"], results_dist[f"X{i}"], color = (0, 0, .5+.5*i/N), label = label, alpha = .5)

plt.plot(results_dist["time"], results_dist["Q"], color = 'lightgreen', label = "Q", lw = 3)
    
plt.plot(results_dist["time"], results_dist["P"], color = 'orange', label = "P")
plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("counts trajectory")
plt.yscale('log')

plt.subplot(212)
all_results_dist = np.array([])
for i in range(N):
    all_results_dist = np.concatenate((all_results_dist, results_dist[f"X{i}"].to_numpy()[int(len(timepoints)/2):]))
    
bins = range(0, int(max(all_results)+1))
plt.hist(R_free["X"].to_numpy()[int(len(timepoints)/2):], bins = bins, density = True, color = 'cyan');

plt.hist(results_dist["Q"].to_numpy()[int(len(timepoints)/2):], bins = bins, density = True, color = 'lightgreen');

plt.hist(all_results_dist, bins = bins, range = (0, bins), density = True, color = 'blue');



meanX_free = np.mean(R_free["X"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanX_free, label = "$<S>_{free}$", ls = "--", color = "gray")

meanX_clamped = np.mean(all_results_dist)
plt.axvline(meanX_clamped, label = "$<S>_{clamped}$", color = "gray")

meanQ = np.mean(results_dist["Q"].to_numpy()[int(len(timepoints)/2):])
plt.axvline(meanQ, label = "$<Q>$", ls = "-.", color = 'gray')

plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("count probability")
plt.tight_layout()
```

# Product Poisson Learning Thermodynamics with Stochastic CRN implementation and N vesicles

```python
#Stochastic Learning wth N Vesicles
k_learn = 1.0
kdb = 10

N = 10 #Number of vesicles
V = N #Large Volume
v = N/V #Vesicle Concentration
X0 = 2

k = k_learn #(k_learn*1.0/V, 1.0*k_learn) #rescale learning rate to match vesicle concentration
delta = .01

#Learning CRN
opposite_side_potential = False

dt = .01
timepoints = np.arange(0, 1000, dt)
max_ind = int(len(timepoints)/10)

species = ["P", "Q", "rxn_Q_P", "rxn_Q_P_r"]
reactions = []
E = {"P":EP, "Q":0}
x0 = {"P":P0, "Q":Q0}
for i in range(N):
    X = "X"+str(i)
    species.append(X)
    E[X]=EX
    x0[X]=X0
    Q = "Q"
    P = "P"
    
    if opposite_side_potential:
        reactions += db_rxns(["P"], [X], E, k = kdb)
    else:
        reactions += db_rxns([], [X, P], E, k = kdb) 
        
    reactions += clamping_reactions(P, Q, X, epsilon = k_learn, k = k, V = V, delta = delta,
                                    opposite_side_potential = opposite_side_potential)
    
    rSf = f"rxn_{X}_{P}"
    rSr = f"rxn_{X}_{P}_r"
    species += [rSf, rSr]

M_pp_N = Model(species = species, reactions = reactions, initial_condition_dict = x0)
results = py_simulate_model(timepoints, M_pp_N, stochastic = True)

plt.figure(figsize = (14, 10))
plt.subplot(411)

title = "\nFree CRN: $\emptyset \leftrightharpoons S$\n"
title += "Clamped CRN: $\emptyset \leftrightharpoons S + P \quad P + S \\rightarrow 2P + S \quad Q + P \\rightarrow Q$"
plt.title(title, fontsize = "xx-large", loc = 'left')
#plt.plot(R_free["time"][:max_ind], R_free["X"][:max_ind], color = 'cyan', label = "S Free")
plt.plot(results["time"][:max_ind], results["Q"][:max_ind], color = 'lightgreen', label = "Q", lw = 3)
for i in range(N):
    if i == N/2:
        label = "S Clamped"
    else:
        label = None
    plt.plot(results["time"][:max_ind], results[f"X{i}"][:max_ind], color = (0, 0, .5+.5*i/N), label = label, alpha = .5)
plt.plot(results["time"][:max_ind], results["P"][:max_ind], color = 'orange', label = "P")
plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("counts trajectory")
plt.yscale('log')


plt.subplot(412)
all_results = np.array([])
for i in range(N):
    all_results = np.concatenate((all_results, results[f"X{i}"][int(len(timepoints)/2):].to_numpy()))

meanX_clamped = np.mean(all_results)
bins = int(max(all_results)+1)
plt.hist(R_free["X"].to_numpy()[int(len(timepoints)/2):], bins = bins, range = (0, bins), density = True, color = 'cyan');
plt.hist(all_results, bins = bins, range = (0, bins), density = True, color = 'blue');

plt.axvline(Q0, lw = 3, color = "lightgreen")

plt.axvline(meanX_free, label = "$<S>_{free}$", ls = "--", color = "gray")
plt.axvline(meanX_clamped, label = "$<S>_{clamped}$", color = "gray")

plt.legend(loc = 1, fontsize = 'large')
plt.ylabel("count probability")
plt.tight_layout()


plt.subplot(413)
for i in range(N):
    if i == int(N/2):
        plt.plot(timepoints[:max_ind], results[f"rxn_X{i}_P"][:max_ind], label = f"rxn_X{i}_P", color = 'blue', alpha = .5)
        plt.plot(timepoints[:max_ind], results[f"rxn_X{i}_P_r"][:max_ind], label = f"rxn_X{i}_P_r", color = 'green', alpha = .5)
    else:
        plt.plot(timepoints[:max_ind], results[f"rxn_X{i}_P"][:max_ind], color = 'blue', alpha = .5)
        plt.plot(timepoints[:max_ind], results[f"rxn_X{i}_P_r"][:max_ind], color = 'green', alpha = .5)

plt.plot(timepoints[:max_ind], results[f"rxn_Q_P"][:max_ind], label = f"rxn_Q_P", color = 'red', alpha = .5)
plt.plot(timepoints[:max_ind], results[f"rxn_Q_P_r"][:max_ind], label = f"rxn_Q_P_r", color = 'purple', alpha = .5)
plt.legend()
plt.yscale('log')


#Compute fluxes
JQ = (results[f"rxn_Q_P"].to_numpy()[1:]-results[f"rxn_Q_P_r"].to_numpy()[:-1])/dt
JS = np.zeros(len(timepoints)-1)
for i in range(N):
    JS += (results[f"rxn_X{i}_P"].to_numpy()[1:]-results[f"rxn_X{i}_P_r"].to_numpy()[:-1])/dt

plt.subplot(414)
#plt.plot(timepoints[:max_ind], JS[:max_ind], label = "Flux S")
#plt.plot(timepoints[:max_ind], JQ[:max_ind], label = "Flux Q")

delta_mean = 100
meanP = np.array([np.mean(results["P"].to_numpy()[i*delta_mean:(i+1)*delta_mean]) for i in range(0, max_ind, delta_mean)])
print(len(meanP))
dSdt = JS[0:max_ind:delta_mean]*np.log(k_learn/delta/meanP)+JQ[0:max_ind:delta_mean]*np.log(k_learn/delta*meanP)
plt.plot(timepoints[0:max_ind:delta_mean], dSdt, label = "Entropy Production")
plt.legend()
```

# Thermodynamics of Learning a Product Poisson with N Vesicles: Hybrid Model

# Computer Thermodynamics for N Poisson Vesicles at various values of Delta

```python
from matplotlib.colors import LogNorm

poisson_thermo_sweep = False

if poisson_thermo_sweep:
    delta_inv = [2, 8, 32, 128, 512, 2048]
    #Q0_list = [10, 100, 1000]
    GS_list = [0, -2, -3, -4, -5, -6, -8]

    Error_HM_emp = np.zeros((len(delta_inv), len(GS_list)))
    Error_HM_thr = np.zeros((len(delta_inv), len(GS_list)))
    Dissipation_Rate_SS_HM = np.zeros((len(delta_inv), len(GS_list)))
    Dissipation_Tot_HM = np.zeros((len(delta_inv), len(GS_list)))

    timepoints = np.arange(0, 10, dt)
    bin_f0= int(4*len(timepoints)/5)

    Q0 = 100
    X0 = 1.0
    P0 = 1.0
    N = 25

    kdb = 500
    dt = .0001
    epsilon = .001

    all_results = {}
    print("Beginning Simulation Loop", end = "...")
    for d_ind, di in enumerate(delta_inv):
        for g_ind, GS in enumerate(GS_list):
            print("di=", di, "Gs=", GS, end = "...")
            delta = epsilon/di

            #Learning CRN
            species = ["P", "Q"]
            E = {"X":GS, "P":EP, "Q":0}
            X0 = int(np.exp(-GS))
            x0 = {"X":X0, "P":P0, "Q":Q0}
            reactions = []
            rules = []
            for i in range(N):
                X = "X"+str(i)
                E[X] = GS
                x0[X] = X0
                species += [X]
                reactions += chemostat_db_rxns([], [X], [], ["P"], E, k = kdb)
                rules += [create_rule_tuple("P", "Q", X, k_learn, dt, delta = delta)]

            np.random.seed()
            M = Model(species = species, reactions = reactions, initial_condition_dict = x0, rules = rules)
            results_thermo = py_simulate_model(timepoints, M, stochastic = True)
                        
            #Compute fluxes
            # J_i^S = [P_i] * <S_i> * (epsilon - delta [P_i])
            JS = np.zeros(len(timepoints))
            for i in range(N):
                JS += results_thermo["P"].to_numpy()*results_thermo[f"X{i}"].to_numpy()*(epsilon - delta*results_thermo["P"].to_numpy())

            # J_i^Q = <Q_i>*(epsilon [P_i] - delta)
            JQ = np.zeros(len(timepoints))
            JQ += results_thermo["Q"].to_numpy()*(epsilon*results_thermo["P"].to_numpy()-delta)

            #T dEntropy_i/dt = RT(J_i^S log( epsilon/(delta P_i)) - J_i^Q log( epsilon P_i / delta ))
            dSdt = np.zeros(len(timepoints))
            dSdt = JS * np.log(epsilon / (delta*results_thermo["P"].to_numpy())) + JQ * np.log(epsilon*results_thermo["P"].to_numpy()/delta)

            #Compute Heatmaps
            mean_Sf = np.mean([np.mean(results_thermo[f"X{i}"][bin_f0:].to_numpy()) for i in range(N)])
            error_f = np.abs(mean_Sf - Q0)/Q0
            Error_HM_emp[d_ind, g_ind] = error_f

            # | delta * (1 - P^2)/(P * (epsilon - delta * P)) |
            mP = np.mean(results_thermo["P"][bin_f0:].to_numpy())
            Error_HM_thr[d_ind, g_ind] = np.abs(delta*(1.0-mP**2.0)/(mP * (epsilon-delta*mP)))

            #Steady State dissipation
            Dissipation_Rate_SS_HM[d_ind, g_ind] = np.mean(dSdt[bin_f0:])

            #Empirical Error Array
            meanX = np.array(np.mean([results_thermo[f"X{i}"].to_numpy() for i in range(N)], 0))
            error = np.abs((meanX - Q0)/Q0)
            #Total Dissipation till steady state
            #Approximate this by when the Empirical Error first reaches the final error calculated in the last bin
            steady_state_start_ind = int(np.argwhere(error < error_f)[0])

            Dissipation_Tot_HM[d_ind, g_ind] = np.sum(dSdt[:steady_state_start_ind])
            all_results[di, GS] = (error, results_thermo["P"].to_numpy(), JS, JQ, dSdt)


            #plt.plot(timepoints, np.mean([results_thermo[f"X{i}"].to_numpy() for i in range(N)], 0), label = f"Gs={GS} di={di}")
    #plt.legend()
    print("Error_HM_emp", Error_HM_emp)
    print("Error_HM_thr", Error_HM_thr)
```

### The following is a simpler example of a Learning CRN as described in the supplemental section of the manuscript

```python
species_Q = ["Q1"]
species_Sc = ["Sc1", "Sc2", "Sc3", "Sc4"]
species_Pc = ["Pc1"]
species_Sf = ["Sf1", "Sf2", "Sf3", "Sf4"]
species_Pf = ["Pf2", "Pf3", "Pf4"]
k_db = 10.0
k_clamp = 0.1
k_learn = 0.1
dt = 0.001

species = species_Q+species_Sc+species_Pc+species_Sf+species_Pf
E = {s:0 for s in species}
reactions = []

def learning_reactions(S, Q, P, k):
    rxns = []
    rxns.append(([Q, P], [Q], "massaction", {"k":k}))
    rxns.append(([S, P], [S, P, P], "massaction", {"k":k}))
    return rxns

reactions += chemostat_db_rxns(["Sc1"], ["Sc2", "Sc3"], ["Pc1"], [], E, k = k_db)
reactions += chemostat_db_rxns(["Sc1", "Sc2"], ["Sc4"], ["Pc1"], [], E, k = k_db)

reactions += chemostat_db_rxns(["Sf1"], ["Sf2", "Sf3"], [],[], E, k = k_db)
reactions += chemostat_db_rxns(["Sf1", "Sf2"], ["Sf4"], [], ["Pf4"], E, k = k_db)

x0 = {
    s : 50 for s in species_Sc+species_Sf
}
x0.update({s : 1.0 for s in species_Pc + species_Pf})

x0["Q1"] = 75

Mf = Model(species = species, reactions = reactions, initial_condition_dict = x0)
M = Model(species = species, reactions = reactions, initial_condition_dict = x0)

add_learning_rule(M, "Pc1", "Q1", "Sc1", k_clamp, dt)
#add_learning_rule(M, "Pf2", "Sc2", "Sf2", k_learn, dt)
#add_learning_rule(M, "Pf3", "Sc3", "Sf3", k_learn, dt)
add_learning_rule(M, "Pf4", "Sc4", "Sf4", k_learn, dt)




timepoints = np.arange(0, 1.0, dt)
Rf = py_simulate_model(timepoints, Mf, stochastic = True, return_dataframe = True)
R = py_simulate_model(timepoints, M, stochastic = True, return_dataframe = True)


plt.figure(figsize = (18, 5))
plt.subplot(141)
plt.title("Unclamped dbCRN", fontsize = 'x-large')
plt.plot(timepoints, Rf["Sc1"], "-.", label = "$S_1$", color = 'blue')
plt.plot(timepoints, Rf["Sc2"], "-.", label = "$S_2$", color = 'red')
plt.plot(timepoints, Rf["Sc3"], "-.", label = "$S_3$", color = 'green')
plt.plot(timepoints, Rf["Sc4"], "-.", label = "$S_4$", color = 'orange')
plt.plot(timepoints, Rf["Sf1"], "--", label = "$S'_1$", color = 'darkblue')
plt.plot(timepoints, Rf["Sf2"], "--", label = "$S'_2$", color = 'darkred')
plt.plot(timepoints, Rf["Sf3"], "--", label = "$S'_3$", color = 'darkgreen')
plt.plot(timepoints, Rf["Sf4"], "--", label = "$S'_4$", color = 'darkorange')
plt.legend(ncol = 2, loc = "upper center")
plt.ylim(-10, 130)
plt.ylabel("Count")
plt.xlabel("time")


plt.subplot(142)
plt.title("$S_1$ is clamped to $Q_1$", fontsize = 'x-large')
plt.plot(timepoints, R["Q1"], "-.", label = "$Q_1$", lw = 2, color = 'gray')
plt.plot(timepoints, R["Sc1"], label = "$S_1$", color = 'blue')
plt.plot(timepoints, R["Sc2"], label = "$S_2$", color = 'red')
plt.plot(timepoints, R["Sc3"], label = "$S_3$", color = 'green')
plt.plot(timepoints, R["Sc4"], label = "$S_4$", color = 'orange')
plt.legend(ncol = 2, loc = "upper center")
plt.ylim(-10, 130)
plt.ylabel("Count")
plt.xlabel("time")

plt.subplot(143)
plt.title("$S'_4$ is clamped to $S_4$", fontsize = 'x-large')
plt.plot(timepoints, R["Q1"], "-.", label = "$Q_1$", lw = 2, color = 'gray')
plt.plot(timepoints, R["Sf1"], label = "$S'_1$", color = 'darkblue')
plt.plot(timepoints, R["Sf2"], label = "$S'_2$", color = 'darkred')
plt.plot(timepoints, R["Sf3"], label = "$S'_3$", color = 'darkgreen')
plt.plot(timepoints, R["Sf4"], label = "$S'_4$", color = 'darkorange')
plt.legend(ncol = 2, loc = "upper center")
plt.ylim(-10, 130)
plt.ylabel("Count")
plt.xlabel("time")

plt.subplot(144)
plt.title("Potential Species During Learning", fontsize = 'x-large')
plt.plot(timepoints, R["Pc1"], label = "$P_1$", color = 'cyan')
plt.plot(timepoints, R["Pf4"], label = "$P'_4$", color = 'gold')
plt.legend()
plt.xlabel("time")
plt.ylabel("Concentration")

plt.savefig("IllustrativeExamplePlot.png")
```
